```python
from envs import BiGlucose, Glucose, CSTR, CartPole
from core.agents import SAC, RLAR, MPCBiGlucose, MPCGlucose, MPCCSTR, MPCCartPole
from IPython.utils import io
```

```python
alt = {'n': 0.2, 'p2': 0.005,  'p3': 5e-6}
env = Glucose(altered_paras=alt)

with io.capture_output() as captured:
    rlar = RLAR(env=env, ckpt_path='saved/RLAR_Glucose.pt')
    sac = SAC(env=env, ckpt_path='saved/SAC_Glucose.pt')
    mpc = MPCGlucose(env=env)

info_rlar = rlar.validate()
info_sac = sac.validate()
info_mpc = mpc.validate()
```
